## Introdução
<img src="aula2-images/principio-dry.png" align="center">
Nesta aula abordaremos técnicas que nos ajudam a ser mais eficientes. O Don't Repeat Yourself (Não Se Repita, em pt-br) é um conceito de programação de computadores o qual propõe que cada porção de conhecimento em um sistema deve possuir uma **representação única**, de autoridade e **livre de ambiguidades** em todo o sistema (WIKIPEDIA). <br>
<br>
DRY significa:
* menos linhas de código = **menos trabalho**
* menos linhas de código = **mais rápido**
* menos linhas de código = **documentação mais fácil** (código mais human-readable)
* menos linhas de código = **felicidade**
* **generalizar** a representação do conhecimento para ser aplicável em contextos similares (porém não iguais)
* ter **cuidado com alterações** na representação do conhecimento (dica: crie testes funcionais e testes unitários)

### Minha primeira função

```python
#Vamos criar uma função que inverte qualquer texto
def inverte_texto(texto):
    #Lembrando a sintaxe de um índice: [begin:end:step]
    #No caso abaixo, dizemos pro python: traga todos os elementos (:) com um passo -1 (:-1)
    print(texto[::-1])
```

```python
inverte_texto("jairo")
```

A função acima imprime um texto, mas frequentemente queremos que o resultado dela seja atribuído a uma variável. Para isso usamos o ***return***

```python
def inverte_texto_com_return(texto):
    #Lembrando a sintaxe de um índice: [begin:end:step]
    #No caso abaixo, dizemos pro python: traga todos os elementos (:) com um passo -1 (:-1)
    return texto[::-1]
```

```python
nome = "jairo"
nome_invertido = inverte_texto_com_return(nome)
print("Meu nome invertido é " + nome_invertido)
```

### Argumentos das funções

Funções suportam múltiplos argumentos. Veja:

```python
#vamos criar uma função que calcula o IMC de uma pessoa
def calc_imc (peso, altura):
    #O IMC é uma medida do peso pela altura ao quadrado
    #Arredondamos o resultado pra facilitar a leitura do usuário
    return round(peso/altura**2) 
```

```python
print(calc_imc(70,1.8)) #A ordem dos parâmetros influencia o resultado
print(calc_imc(1.8, 70))
```

Ao contrário de outras linguagens, Python permite que você passe parâmetros fora de ordem, desde que referencie qual parâmetro está fornecendo.

```python
print(calc_imc(peso=70,altura=1.8))
print(calc_imc(altura=1.8, peso=70)) #Resultados iguais UHUUL!
```

### Valores padrão para argumentos
É muito frequente queremos que uma função tenha vários **valores padrão para parâmetros**. Estes padrões estão lá para facilitar a vida do usuário. Em situações específicas, ele precisará apenas um ou outro parâmetro, e não todos.

```python
#Uma função que separa os elementos de uma string pelo separador
def ler_txt(txt, separador=";"):
    #Inicializa um vetor vazio para armazenar as linhas separadas
    texto_separado = []
    #Para cada linha do txt
    for linha in txt.split("\n"):
        #Insira no final do vetor texto_separado a linha separada pelo delimitador separador
        texto_separado.append(linha.split(separador))
    #Ao fim do processamento, retorne para o usuário a matriz (vetor de vetores) texto_separado
    return texto_separado

#Uma função para imprimir cada linha de uma matriz em uma linha da saída padrão
def imprime_matriz(matriz):
    #Para cada linha na matriz
    for linha in matriz:
        #Imprima esta linha
        print(linha)
```

```python
meu_txt = "Nome;Idade;Apelido\nJairo;22;Jairon\nRodrigo;24;Rodriguinho\nLaryssa;20;Lary"
print(meu_txt)
imprime_matriz(ler_txt(meu_txt))
```

```python
#Neste caso nosso separador é uma tabulação, por isso teremos que configurar na chamada da função este parâmetro
meu_txt = "Nome\tIdade\tApelido\nJairo\t22\tJairon\nRodrigo\t24\tRodriguinho\nLaryssa\t20\tLary"
print(meu_txt)
imprime_matriz(ler_txt(meu_txt, "\t"))
```

### Argumentos opcionais

Python é umas linguagens mais fáceis de lidar com argumentos opcionais. Para indicar que a função poderá receber uma lista de argumentos utilize **\*args**. Caso queira passar uma lista de pares chave=valor, use **\*\*kwargs**.

```python
#Criaremos uma função que soma n elementos, desde que n>0
def soma(num0, *numeros):
    #inicio o resultado como o primeiro valor passado
    resultado = num0
    #caso o usuários tenha fornecido mais parâmetros
    if len(numeros)>0:
        #somo estes parâmetros e armazeno na variável resultado
        for num in numeros:
            resultado += num
    #o resultado é retornado
    return resultado
```

```python
soma(0,1,2,3,4,5)
```

```python
#Suponha que queiramso criar uma landing page com pré-cadastro
#Quanto mais informações o usuário fornecer, melhor
#Mas o único campo obrigatório é o email
#Vamos simular como seria a criação de uma linha para envio pra um arquivo JSON
def pre_cadastro(email,**mais_info):
    #Criamos um registro_bd com apenas o e-mail
    registro_bd = {email : {}}
    #Se houver mais chaves=valores, então adicionaremos estes valores ao registro_bd
    if len(mais_info)>0:
        registro_bd[email].update(mais_info)
    return registro_bd    
```

```python
print(pre_cadastro("jairo@dominio.com"))
print(pre_cadastro("jairo@dominio.com", idade=22, telefone="(11)XXXXX-XXXX"))
```

### Funções anônimas e map\(\)

É comum querermos inserir uma função simples como argumento de algum método no nosso código. Nestes casos queremos que esta função exista apenas naquele momento e não nos preocupamos em chamá-la novamente no futuro. As funções anônimas são isso. 

```python
#Vamos criar uma função não anônima que verifica se uma pessoa é maior de idade
def maioridade(pessoas):
    maioridade = []
    for pessoa in pessoas:
        if pessoa[1] < 18:
            maioridade.append(True)
        else:
            maioridade.append(False)
    return maioridade
```

```python
base = [["Jairo",22,"Anl"],["Laryssa",17,"Estag"],["Bruno", 40, "Dir"]]

print(maioridade(base))
```

Para criar uma função anônima, utilizaremos a função ***map()***. Esta função aplica uma função a cada membro de um dado iterável e retorna o resultado. Geralmente a função aplicada dentro do map() é uma função anônima.<br>
map() retorna um objeto tipo map, por isso no exemplo converterei o resultado para vetor usando list().

```python
#imprima a lista resultante da avaliação elemento a elemento da base
#avalie como True se o segundo elemento da linha pessoa[1] for menor que 18
#e falo caso contrário
print(list(map(lambda pessoa : True if pessoa[1] < 18 else False, base)))
```

### Doctest
Você deve se lembrar que passa a maior parte do seu tempo fazendo manutenção em seu código. Por isso, vale a pena criar testes unitários em todo ele para garantir que cada parte está se comportando como previsto após uma alteração.

```python
import doctest
```

```python
def pre_cadastro(email,**mais_info):
    #Estas linhas com >>> simulam a chamada da função
    #O doctest compara a saída com a linha seguinte
    """
    >>> pre_cadastro("jairo@dominio.com")
    {'jairo@dominio.com': {}}
    >>> pre_cadastro("jairo@dominio.com", idade=22, telefone="(11)XXXXX-XXXX")
    {'jairo@dominio.com': {'idade': 22, 'telefone': '(11)XXXXX-XXXX'}}
    """
    #Criamos um registro_bd com apenas o e-mail
    registro_bd = {email : {}}
    #Se houver mais chaves=valores, então adicionaremos estes valores ao registro_bd
    if len(mais_info)>0:
        registro_bd[email].update(mais_info)
    return registro_bd

#abre o arquivo log.txt no modo append
f = open('log.txt', 'a')
#escreve no arquivo o resultado do testmod()
f.write('LOG pre_cadastro:\t' + str(doctest.testmod())+'\n')
f.close()
```

### Getpass
Getpass é uma biblioteca para entrada de senha. Possui duas funções:
* **getpass**: Abre um campo no console pro usuário digitar sua senha, colocando asterixos no lugar das letras digitadas.
* **getuser**: Esta função busca a primeira ocorrência não vazia das variáveis LOGNAME, USER, LNAME e USERNAME.

```python
import getpass
```

```python
def login(login=getpass.getuser()):
    senha = getpass.getpass(prompt="Senha:")
    print("Usuário:"+str(login))
    print("Senha:"+str(senha))
```

```python
login()
login("teste")
```

### Urllib + BeautifulSoup
Um bem casado para extração (data mining) de dados estruturados na web.

```python
import urllib #gerencia requisições http
#Quando você deseja importar apenas um pedaço de uma biblioteca
#você pode usar from <nome_da_biblioteca> import <nome_do_modulo>
from bs4 import BeautifulSoup #processa documentos html e xml
```

```python
#Requisitamos o conteúdo de uma página
conn = urllib.request.urlopen('http://www.python.org')
#Lemos o resultado desta requisição
return_str = conn.read()
```

```python
#Veja o HTML da página
print(return_str)
```

```python
#Para trabalhar melhor com o HTML, usamos BeautifulSoup
soup = BeautifulSoup(return_str)
print(soup)
```

```python
#Suponha que queremos saber qual o título da página
print(soup.title)
```

```python
#Se preferir apenas o conteúdo em texto entre as tags, chame o método string
print(soup.title.string)
```

```python
#É muito comum listar todos os links de uma página
#O código abaixo obtém o href de todas as tags a da página
for link in soup.find_all("a"):
    print(link.get('href'))
```

```python
#O mesmo poderia ser feito via função anônima
print(list(map(lambda link : link.get('href'), soup.find_all("a"))))
```

```python
#Como temos muitos links que não levam a páginas de fato
#(são apenas eventos javascript ou links pra mesma página)
#vamos selecionar apenas links que comecem com uma barra ou http
def scrapper_links_validos(soup):
    links_validos=[]
    #para cada link da página
    for link in soup.find_all("a"):
        #verifique se começa com "/" ou "http"
        if link.get('href').startswith("/") or link.get('href').startswith("http"):
            #se for o caso, adicione ao vetor links_validos
            links_validos.append(link.get('href'))
    return links_validos        
```

```python
print(scrapper_links_validos(soup))
```
